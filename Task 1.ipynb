```python
country_by_1960 = df.sort_values(by='1960').head(20)
country_by_1960
```

```python
country_by_1960_t = df.transpose()
countries = country_by_1960_t.index
years = country_by_1960_t.columns

# Plotting each country's data values separately
for i, country in enumerate(countries):
    plt.figure(figsize=(10, 6))  # Adjust figsize as needed
    plt.bar(years, country_by_1960_t.loc[country], color='blue')  # Change color if needed
    plt.xlabel('Year')
    plt.ylabel('Data Values')
    plt.title(f'Data Values for {country}')
    plt.xticks(rotation=70)
    plt.show()
```

```python
df_melted = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                    var_name='Year', 
                    value_name='Population')
```

```python
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
df_filtered = df_melted[(df_melted['Year'] >= 1960) & (df_melted['Year'] <= 2023)]
```

```python
df_filtered = df_filtered.dropna(subset=['Population'])
```

```python
avg_population = df_filtered.groupby('Country Name')['Population'].mean().reset_index()

# histogram for the distribution of average population values from 1960 to 2023
plt.figure(figsize=(10, 6))
plt.hist(avg_population['Population'], bins=20, edgecolor='black', color='skyblue')
plt.title('Distribution of Average Population Values (1960-2023)')
plt.xlabel('Average Population')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```
