```python
high_corr = ['emp.var.rate','euribor3m','nr.employed']
```

```python
data1 = task3.copy()
data1.columns
```

```python
data1.drop(high_corr,inplace=True,axis=1)  
data1.columns
```

```python
data1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
encoded_val = data1.apply(lb.fit_transform)
encoded_val
```

```python
encoded_val['deposit'].value_counts()
```

```python
x = encoded_val.drop('deposit',axis=1) 
y = encoded_val['deposit']             
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

print(41188*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
